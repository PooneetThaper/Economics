```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Part 1: Table of values

```python
Qd = np.arange(100,-1,-12.5)
Qs = np.arange(10,100,10)
P = [0,2.5,5,7.5,10,12.5,15,17.5,20]
df = pd.DataFrame([[P[i],Qd[i],Qs[i]] for i in range(9)],
        columns=['Price', 'Quantity Demanded', 'Quantity Supplied'])
print(df)
```

## Part 2: Graph quantities supplied and demanded

```python
sns.set()
_ = plt.plot(df['Quantity Demanded'], df['Price'], marker='.')
_ = plt.plot(df['Quantity Supplied'], df['Price'], marker='.')
_ = plt.xlim(0,105)
_ = plt.ylim(0,25)
_ = plt.xticks(np.arange(0,101,10))
_ = plt.yticks(np.arange(0,25,2.5))
_ = plt.hlines(10,0,50,linestyles='dashed')
_ = plt.vlines(50,0,10,linestyles='dashed')
_ = plt.xlabel('Quantity')
_ = plt.ylabel('Price')
_ = plt.title('Amusement Park Pass Market')
_ = plt.legend(["Qd", "Qs"])

plt.show()
```
